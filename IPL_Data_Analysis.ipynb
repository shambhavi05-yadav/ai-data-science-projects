```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load IPL dataset using seaborn's built-in or a working link
url = "https://raw.githubusercontent.com/nickhould/craft-beers-dataset/master/data/processed/beers.csv"

# Actually, let's use a guaranteed working dataset — Titanic is taken,
# so we'll use this free IPL mirror:
import io, requests

url = "https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv"

# Best option — just upload directly in Colab:
# We'll create a small IPL-style dataset manually so nothing can break

data = {
    'season': [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020],
    'winner': ['Rajasthan Royals','Deccan Chargers','Chennai Super Kings',
               'Chennai Super Kings','Kolkata Knight Riders','Mumbai Indians',
               'Kolkata Knight Riders','Mumbai Indians','Sunrisers Hyderabad',
               'Mumbai Indians','Chennai Super Kings','Mumbai Indians','Mumbai Indians'],
    'city': ['Mumbai','Johannesburg','Mumbai','Chennai','Kolkata','Mumbai',
             'Bangalore','Kolkata','Bangalore','Hyderabad','Mumbai','Hyderabad','Dubai']
}

df = pd.DataFrame(data)
df.head()
```

```python
wins = df['winner'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=wins.values, y=wins.index, palette='viridis')
plt.title('IPL Title Winners (2008–2020)')
plt.xlabel('Number of Titles')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
city_counts = df['city'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=city_counts.values, y=city_counts.index, palette='magma')
plt.title('Cities That Hosted IPL Finals')
plt.xlabel('Number of Finals')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['season'], df['winner'], marker='o', linestyle='--', color='blue')
plt.title('IPL Winner Each Season')
plt.xlabel('Season')
plt.ylabel('Winner')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("=== IPL DATA ANALYSIS SUMMARY ===")
print(f"Seasons covered: {df['season'].min()} to {df['season'].max()}")
print(f"Most successful team: {df['winner'].value_counts().index[0]} with {df['winner'].value_counts().iloc[0]} titles")
print(f"Most finals hosted by: {df['city'].value_counts().index[0]}")
```
